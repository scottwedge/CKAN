```python
import requests
import pandas as pd

resource_id = "6d265cf0-ebea-4d6b-9ddf-524739f750fe"

resultslimit = 500
county = "Albany"
#characteristicname = ["Phosphorus","Nitrogen"]

params={
    'resource_id': resource_id , 
    'fields' : "LakeDatabase 2::County,Characteristic_Name,Result Value,SAMPLE_DATE",
    'limit' : resultslimit,
    'filters' : '{"Characteristic_Name": ["Phosphorus","Nitrogen"],"LakeDatabase 2::County": ["Albany", "Chautauqua"]}',
}


url = 'http://saturn04.ctg.albany.edu/api/3/action/datastore_search'
results = requests.get(url, params=params).json()
print (results)
```

```python
df = pd.DataFrame(results['result']['records'])
df.head(50)
```

```python
df.shape
```

```python
df.rename(columns={'LakeDatabase 2::County':'county',
                          'Result Value':'Result_value',}, 
                 inplace=True)
df.head()
```

```python
#Converting Sample Dates to pandas datetime format and define the resultvalue dtype
df['SAMPLE_DATE'] = pd.to_datetime(df.SAMPLE_DATE)
df["Result_value"] = pd.to_numeric(df.Result_value)
df.sort_values(by='SAMPLE_DATE',inplace=True)
df.head(10)
```

```python
#Importing the matplotlib library
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
df_plot = df.loc[(df.county == 'Albany') & (df.Characteristic_Name == 'Phosphorus')]
#dfplot.shape
#Plot a series chart by taking ResultValue mean for each year
 
phosphorus_line_chart = df_plot.groupby(df_plot['SAMPLE_DATE'].dt.year)['Result_value'].mean().plot(title='Albany Phosporus yearly Analysis')

#phosphorus_line_chart = phosphorus_line_chart.loc[phosphorus_line_chart['WATER_NAME'] == 'Chautauqua Lake']
phosphorus_line_chart.legend()
df_plot.head()

```

```python
# gca stands for 'get current axis'
ax = plt.gca()

df_plot.plot(kind='line',x='SAMPLE_DATE',y='Result_value',ax=ax)
#df.plot(kind='line',x='Date',y='num_pets', color='red', ax=ax)

plt.show()
```
